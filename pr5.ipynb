```python
import pandas as pd
import matplotlib.pyplot as plt
table = pd.read_csv("ggger.csv")
table
```

Определю рейтинг в разных годах через Линейный график

```python
plt.plot(table['Year'], table['Rating'])
plt.xlabel('Год')
plt.ylabel('Рейтинг')
plt.title('Рейтинг в разные года')
plt.grid()
plt.legend()
plt.show()
```

Minifigures и Current_Price

```python
plt.plot(table['Minifigures'], table['Current_Price'])
plt.xlabel('Год')
plt.ylabel('Название')
plt.title('Рейтинг в разные года')
plt.grid()
plt.legend()
plt.show()
```

Piece и Rating

```python
plt.bar(table['Minifigures'], table['Current_Price'], color="pink")
plt.xlabel('Год')
plt.ylabel('Название')
plt.title('Рейтинг в разные года')
plt.legend()
plt.show()
```

Year и Theme_Group

```python
tablenew = table.drop_duplicates(subset=['Year', 'Rating', 'Theme_Group'])

plt.pie(tablenew['Year'], autopct='%1.1f%%')
plt.title('Theme_Group')
plt.show()
```

Rating

```python
plt.pie(tablenew['Rating'], labels=tablenew['Theme_Group'], autopct='%1.1f%%')
plt.title('Theme_Group')
plt.show()
```

Year	Pieces	Name dвыборка данных
